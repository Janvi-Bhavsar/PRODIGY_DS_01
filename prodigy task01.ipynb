```python
# URL of the dataset
url = 'https://raw.githubusercontent.com/datasets/population/master/data/population.csv'

# Load the dataset
data = pd.read_csv(url)
```

```python
# Inspect the first few rows of the dataset
print(data.head())
# Inspect the column names
print(data.columns)
```

```python
# Select relevant columns
population_data = data[['Year', 'Value']]
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(population_data['Year'], population_data['Value'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Global Population Over Years')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a histogram of the population values
plt.figure(figsize=(12, 6))
plt.hist(population_data['Value'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population Values')
plt.show()
```
